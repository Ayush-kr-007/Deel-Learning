```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Single Layer Perceptron Dataset.csv')
```

```python
print(df.shape)
df.head()
```

```python
sns.scatterplot(x=df['Feature2'], y=df['Feature3'], hue=df['Class_Label'])
```

```python
df = df.drop(['Pattern','Feature1'],axis=1)
```

```python
df = df.dropna()
y = df['Class_Label']
X = df.drop(['Class_Label'],axis=1)
```

```python
from sklearn.linear_model import Perceptron
p = Perceptron()
```

```python
p.fit(X,y)
```

```python
p.coef_
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X.values, y.values.astype(np.int_), clf=p, legend=2)
```
